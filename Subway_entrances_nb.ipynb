```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import geopandas as gpd
%matplotlib inline

pd.set_option("max_rows",100)
```

```python
# Filepath to read Subway entrances shapefile
filepath_subway_entrances = "Subway Entrances Shapefiles/geo_export_485d1373-2cdc-4b7b-a550-584bb82e0d6f.shp"
filepath_nyc_boundaries = "Borough Boundaries/geo_export_df017658-06e6-4fbf-8be1-4070c446823f.shp"
```

```python
Subway_gdf = gpd.read_file(filepath_subway_entrances)
NYC_gdf = gpd.read_file(filepath_nyc_boundaries)
```
